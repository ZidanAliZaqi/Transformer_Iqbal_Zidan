```python
from google.colab import files
from transformers import BartForConditionalGeneration, BartTokenizer

# 1. Upload the document file
uploaded = files.upload()

# 2. Load Model dan Tokenizer
model_name = "facebook/bart-large-cnn"
model = BartForConditionalGeneration.from_pretrained(model_name)
tokenizer = BartTokenizer.from_pretrained(model_name)

# 3. Read the uploaded document with error handling for different encodings
filename = list(uploaded.keys())[0]
try:
    with open(filename, 'r', encoding='utf-8') as file:
        document = file.read()
except UnicodeDecodeError:
    with open(filename, 'r', encoding='ISO-8859-1') as file:
        document = file.read()

# 4. Preprocessing Input
inputs = tokenizer.encode("summarize: " + document, return_tensors="pt", max_length=1024, truncation=True)

# 5. Generate Summary with approximately 200 words
summary_ids = model.generate(
    inputs,
    max_length=260,  # Adjusted to allow around 200 words
    min_length=150,  # Ensures minimum length is reasonable
    length_penalty=2.0,
    num_beams=4,
    early_stopping=True
)

# 6. Decode Output
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

# 7. Output the document and its summary
print("Original Document:")
print(document)
print("\nSummary:")
print(summary)
```
